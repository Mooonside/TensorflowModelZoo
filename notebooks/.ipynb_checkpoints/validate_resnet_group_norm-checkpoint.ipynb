```python
import tensorflow as tf
import numpy as np
import sys
import os
import cv2
import matplotlib.pyplot as plt

os.environ['CUDA_VISIBLE_DEVICES'] = '1'
sys.path.insert(0, '..')
arg_scope = tf.contrib.framework.arg_scope


%matplotlib inline
%load_ext autoreload
% autoreload 2
```

```python
from networks.resnet import resnet_arg_scope_gn, resnet_v1_101, arg_scope
```

```python
inputs = tf.placeholder(shape=[None, 224, 224, 3], name='inputs', dtype=tf.float32)
```

```python
with arg_scope(resnet_arg_scope_gn(weight_decay=1e-4, group_norm_num=32)):
    net, end_points = resnet_v1_101(inputs, num_classes=1000, is_training=True, scope='resnet_v1_101_gn')
```

```python
ckpt_path = '/home/chenyifeng/TF_Models/ptrain/ILSVRC/ResNet_101_V1_GN/model.ckpt'
sess = tf.Session()
saver = tf.train.Saver()
saver.restore(sess, ckpt_path)
```

```python
def ilsvrc_preprocess(img):
    img = img.astype(np.float32)
    img -= np.array([[[102.9801, 115.9465, 122.7717]]])
    
    img_h, img_w = img.shape[0:2]
    img = cv2.resize(img, (224, 224), interpolation=cv2.INTER_LINEAR)
    # pad img at right bottom so that it can be divided by stride
    img = np.expand_dims(img, axis=0)
    return img
```

```python
pics_path = '/mnt/ftp/datasets/ImageNet/ISVRC2015/challenge/dataset/CLS-LOC/ILSVRC2015_CLS-LOC/Data/CLS-LOC/train/n04141975'
pic_name = os.listdir(pics_path)[9]
pic_path = os.path.join(pics_path, pic_name)

img = cv2.imread(pic_path)
plt.figure(figsize=(5, 5))
plt.imshow(img[:, :, ::-1])
# keep input in bgr order!
net_v = sess.run(net, feed_dict={inputs:ilsvrc_preprocess(img)})

net_v = np.squeeze(net_v)
print(np.argmax(net_v))
```
